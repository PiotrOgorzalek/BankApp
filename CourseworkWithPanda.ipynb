```
import pandas as pd
import swifter
import random
#function change integers to words
def change(text):
  index = imdb.get_word_index()
  reverse_index = dict([(value, key) for (key, value) in index.items()])
  text2 = " ".join( [reverse_index.get(i - 3, "#") for i in text ])
  return text2
#df = pd.DataFrame.from_records(training_data,training_targets)
#creating data frame
dataset = pd.DataFrame()
datasetTest = pd.DataFrame()
num1 = random.randrange(23000)
num2 = num1+2000

#decoded = " ".join( [reverse_index.get(i - 3, "#") for i in training_data[1]] )
##create training data set 
dataset["sentences"] = training_data[num1:num2].tolist()
dataset["target"] = training_targets[num1:num2].tolist()
#that line gives me the most issues when applied need to change training data to 1000 to get some normal exexcution time
dataset['sentences'] = dataset.sentences.swifter.apply(change)
datasetTest["sen"] = testing_data[num1:num2].tolist()
datasetTest["target"] = testing_targets[num1:num2].tolist()
datasetTest["sen"] = datasetTest.sen.swifter.apply(change)
dataset
```

```
!pip install texthero
import texthero as hero
from texthero import stopwords
from texthero import preprocessing
#removing stop words
default_stopwords = stopwords.DEFAULT
##had to add br to stopword as it was showing a lot of bigrams with it
custom_stopwords = default_stopwords.union(set(['br']))

dataset['sentences'] = hero.remove_stopwords(dataset['sentences'],custom_stopwords)
dataset['sentences'] = hero.remove_whitespace(dataset['sentences'])
datasetTest['sen'] = hero.remove_stopwords(datasetTest['sen'],custom_stopwords)
datasetTest['sen'] = hero.remove_whitespace(datasetTest['sen'])
```

creating bigrams 

```
from nltk.util import ngrams
import nltk
nltk.download('punkt')
from nltk import word_tokenize
#getting sentences

sentences_only = dataset["sentences"]
sentences_only1 = datasetTest["sen"]
def extract_bigrams(sentences): 
 all_bigrams = []
 
 for sentence in sentences:
   token = word_tokenize(sentence)
   bigrams = ngrams(token,2)
   for b in bigrams:
     all_bigrams.append(b)
 return all_bigrams
 
extracted_bigrams = extract_bigrams(sentences_only)
extracted_bigrams1= extract_bigrams(sentences_only1)
```

```
sentences_only = dataset["sentences"]

def extract_unigrams(sentences): 
 all_unigrams = []
 for sentence in sentences:
   token = word_tokenize(sentence)
   #token = [word for word in sentence if not word in stopwords.words('english')]
  # filtered = (" ").join(token)
   unigrams = ngrams(token,1)
   for b in unigrams:
     all_unigrams.append(b)
 return all_unigrams
 
extracted_unigrams = extract_unigrams(sentences_only)
extracted_unigrams1 = extract_unigrams(datasetTest["sen"])
```
